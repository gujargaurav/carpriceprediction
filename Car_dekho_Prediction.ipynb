```python
final_df=df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven','Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
final_df['Current_year']=2020
```

```python
final_df.head()
```

```python
final_df['No_of_years']=final_df['Current_year']- final_df['Year']
```

```python
final_df.head()
```

```python
final_df.drop(['Year'],axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df.drop(['Current_year'],axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_set=pd.get_dummies(final_df,drop_first=True)
```

```python
#independent and dependent features
X=final_set.iloc[:,1:]
y=final_set.iloc[:,0]
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model=ExtraTreesRegressor()
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
#plot graph of feature importances for better visualization 
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(5).plot(kind='barh') 
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_random = RandomForestRegressor(random_state=1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
predictions=rf_random.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test,predictions)
```

```python
print(f"Mean absolute Error : {mae}")
```

```python
sns.scatterplot(y_test,predictions)
```

```python
sns.distplot(y_test-predictions)
```
